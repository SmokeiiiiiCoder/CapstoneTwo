# Exploratory Data Analysis (EDA)


```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#import the data
path_s = './data/supplemental_files/'
path_t = './data/train_files/'
path_e = './data/example_test_files/'
```

```python
prices1 = pd.read_csv(f'{path_t}stock_prices.csv')
prices2 = pd.read_csv(f'{path_t}secondary_stock_prices.csv')

prices1.shape , prices2.shape
```

```python
#combine all data together
df = pd.concat([prices1,prices2])
df
```

```python
#getting more information to understand the data
df.info()
```

```python
#calculate high-low as performance in one day
df['Performance'] = (df['High'] - df['Low'])/df['Volume']
```

```python
corr = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
hm = sns.heatmap(corr, annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
```
